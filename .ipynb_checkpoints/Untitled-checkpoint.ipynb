```python

import face_recognition
import numpy as np
import os
from PIL import Image
from flask import Flask, request, jsonify
import io

app = Flask(__name__)
KNOWN_DIR = "known_faces"
EMBED_DIR = "embeddings"

def save_embedding(image_path, name):
    image = face_recognition.load_image_file(image_path)
    encodings = face_recognition.face_encodings(image)
    if encodings:
        np.save(os.path.join(EMBED_DIR, f"{name}.npy"), encodings[0])
        print(f"Embedding enregistré pour {name}")
    else:
        print(f"Aucun visage détecté pour {name}")

def load_known_faces():
    known_faces = {}
    for file in os.listdir(EMBED_DIR):
        if file.endswith(".npy"):
            name = file.replace(".npy", "")
            encoding = np.load(os.path.join(EMBED_DIR, file))
            known_faces[name] = encoding
    return known_faces

@app.route("/recognize", methods=["POST"])
def recognize():
    try:
        image_stream = io.BytesIO(request.data)
        img = Image.open(image_stream).convert("RGB")
        img_array = np.array(img)

        input_encodings = face_recognition.face_encodings(img_array)
        if not input_encodings:
            return jsonify({"status": "error", "message": "Aucun visage détecté"})

        input_encoding = input_encodings[0]
        known_faces = load_known_faces()

        for name, encoding in known_faces.items():
            matches = face_recognition.compare_faces([encoding], input_encoding, tolerance=0.5)
            if matches[0]:
                return jsonify({"status": "success", "identity": name})

        return jsonify({"status": "success", "identity": "inconnu"})

    except Exception as e:
        return jsonify({"status": "error", "message": str(e)})

if __name__ == "__main__":
    os.makedirs(KNOWN_DIR, exist_ok=True)
    os.makedirs(EMBED_DIR, exist_ok=True)
    # Exemple d'enregistrement : save_embedding('known_faces/ahmed.jpg', 'ahmed')
    app.run(host="0.0.0.0", port=5000)
```
